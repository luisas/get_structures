```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import glob
import matplotlib.pyplot as plt
sns.set_context("talk")
```

```python
hits_path = "/home/luisasantus/Desktop/crg_cluster/projects/nf-core-getstructures/outdir_secondrun/mmseqs/search_output/*m8"
```

```python
hits_summary = pd.DataFrame()
for file in glob.glob(hits_path):
    family = os.path.splitext(os.path.basename(file))[0]
    if os.path.getsize(file) == 0:
        print(file)
    if os.path.getsize(file) > 0: 
        df = pd.read_csv(file, sep= "\t", header = None)
        df["family"] = family
        hits_summary = pd.concat([hits_summary, df])
    
```
